```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data=pd.read_csv('D:\\PIAIC DOC\\Q2\\Machine Learning\\YouTube-ML\\DataSets\\Position_Salaries.csv')
```

# Spliting into Input and Output

```python
X=data.iloc[: , 1].values
Y=data.iloc[: , 2].values

X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
```

# Linear Regressor

```python
LR=LinearRegression()
```

# Training a model

```python
LR.fit(X,Y)
```

# Accuracy of the linear Regression Model

```python
Acc=LR.score(X,Y)
print("Accuracy of Linear REgression Model : " , Acc )
```

# Polynomial Regressor

```python
PR=PolynomialFeatures(4)
```

# Fit Tranform of X

```python
poly_x=PR.fit_transform(X)
```

# Training a model

```python
PR.fit(poly_x,Y)
```

# Linear Regressor for Polynomial

```python
LR2=LinearRegression()
LR2.fit(poly_x,Y)
```

# Accuracy of the Polynomial Regression Model

```python
Acc=LR2.score(poly_x,Y)
print("Accuracy of Polynomial Linear REgression Model : " , Acc*100 )
```

# Visualization by graph (Linear Regression)

```python
plt.figure(figsize=(10,6))
plt.title("Salary With Position Level",fontsize=18)
plt.xlabel("Position Level of the Employees",fontsize=15)
plt.ylabel("Salary of the Employees",fontsize=15)
plt.scatter(X,Y)
plt.plot(X,LR.predict(X))
plt.show()
```

# Visualization by graph (Polynomial Linear Regression)

```python
plt.figure(figsize=(10,6))
plt.title("Salary With Position Level",fontsize=18)
plt.xlabel("Position Level of the Employees",fontsize=15)
plt.ylabel("Salary of the Employees",fontsize=15)
plt.scatter(X,Y,color='#0da6c1')
plt.plot(X,LR2.predict(PR.fit_transform(X)),color='#68943b')
plt.show()
```
