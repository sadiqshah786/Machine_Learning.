```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data=data=pd.read_csv('D:\PIAIC DOC\Q2\Machine Learning\YouTube-ML\DataSets\salaryData.csv')
```

# Splitng into Input and Output 

```python
X=data.iloc[: , 0].values.reshape(-1,1)
Y=data.iloc[: , 1].values.reshape(-1,1)
```

# Linear Regressor

```python
LR=LinearRegression()
```

# Training a Model

```python
LR.fit(X,Y)
```

# Accuracy of the Model

```python
Acc=LR.score(X,Y)*100
print("Accurary of the Model : " ,Acc,"%")
```

# Visualization by graph (Linear Regression)

```python
plt.figure(figsize=(10,6))
plt.title("(Year of Experience with Salary)",fontsize=18)
plt.xlabel("Experience of the Employees",fontsize=15)
plt.ylabel("Salary of the Employees",fontsize=15)
plt.scatter(X,Y)
plt.plot(X,LR.predict(X))
```

# Polynomial Regressor

```python
PLR=PolynomialFeatures(degree=15)
```

## Fit Tranform X

```python
Poly_x=PLR.fit_transform(X)
PLR.fit(Poly_x,Y)
```

# Linear Regressor for Polynomial

```python
LR2=LinearRegression()
```

# Traning a Model

```python
LR2.fit(Poly_x,Y)
```

# Accuracy of the Model

```python
Acc=LR2.score(Poly_x,Y)
print("Accuracy of Polynomial Linear REgression Model : " , Acc*100 )
```

# Visualization by graph (Polynomial Linear Regression)

```python
plt.figure(figsize=(10,6))
plt.title("(Year of Experience with Salary)",fontsize=18)
plt.xlabel("Experience of the Employees",fontsize=15)
plt.ylabel("Salary of the Employees",fontsize=15)
plt.scatter(X,Y,color='#0da6c1')
plt.plot(X,LR2.predict(PLR.fit_transform(X)),color='#68943b')
plt.show()
```
